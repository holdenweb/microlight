### Switching Response

Firstly I wanted to plot the voltages on Q1's collector and Q2's emitter
against the input voltage to verify switching operation.

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
data = [
    # Vin, V1, V2 in Figure 1
    (0.0, 11.83, 11.92),
    (0.33, 11.83, 11.92),
    (0.4, 11.83, 11.92),
    (0.45, 11.83, 11.92),
    (0.5, 11.83, 11.92),
    (0.55, 11.12, 11.82),
    (0.6, 6.91, 7.58),
    (0.65, 0.02, 0.66),
    (0.7, 0.01, 0.63),
    (3.3, 0.01, 0.63)
]
```

```python
vin, v1, v2 = zip(*data)
```

```python
plt.plot(vin, v1, label='$Q_1C$')
plt.plot(vin, v2, label='$Q_2E$')
plt.legend(loc="upper right")
```

This looks good. Switching occurs very shortly after the maximum zero logic level is exceeded
so switching should be reliable.
